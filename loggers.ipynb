```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime
```

```python
df_dunajec = pd.read_csv("csv/dunajec.csv")
df_dunajec['date_time'] = pd.to_datetime(df_dunajec['date_time'], format='%m/%d/%Y %H:%M')
df_dunajec.set_index('date_time', inplace=True)
df_dunajec_resampled = df_dunajec.resample('D').mean()

df_31 = pd.read_csv("csv/S31.csv")
df_31['date_time'] = pd.to_datetime(df_31['date_time'], format='%m/%d/%Y %H:%M')
df_31.set_index('date_time', inplace=True)
df_31_resampled = df_31.resample('D').mean()

df_36 = pd.read_csv("csv/S36.csv")
df_36['date_time'] = pd.to_datetime(df_36['date_time'], format='%m/%d/%Y %H:%M')
df_36.set_index('date_time', inplace=True)
df_36_resampled = df_36.resample('D').mean()

df_37 = pd.read_csv("csv/S37.csv")
df_37['date_time'] = pd.to_datetime(df_37['date_time'], format='%m/%d/%Y %H:%M')
df_37.set_index('date_time', inplace=True)
df_37_resampled = df_37.resample('D').mean()

df_39 = pd.read_csv("csv/S39.csv")
df_39['date_time'] = pd.to_datetime(df_39['date_time'], format='%m/%d/%Y %H:%M')
df_39.set_index('date_time', inplace=True)
df_39_resampled = df_39.resample('D').mean()

df_rembud = pd.read_csv("csv/rembud.csv")
df_rembud['date_time'] = pd.to_datetime(df_rembud['date_time'], format='%m/%d/%Y %H:%M')
df_rembud.set_index('date_time', inplace=True)
df_rembud_resampled = df_rembud.resample('D').mean()
```

```python
date_36 = df_36.index
gw_36 = df_36['depth_to_gw']
temp_36 = df_36['temp']
ec_36 = df_36['ec']

fig, ax1 = plt.subplots(figsize=(17, 7))
ax2 = ax1.twinx()

ax1.plot(date_36, gw_36, "g", label="Depth to water level", alpha=0.75)
ax2.plot(date_36, ec_36, "b", label="EC", alpha=0.75)

ax1.invert_yaxis()

for ax in ax1, ax2:
    ax.grid(False)

ax1.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
ax1.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
start_date = datetime.strptime("2022-08-15", "%Y-%m-%d")
end_date = datetime.strptime("2023-10-15", "%Y-%m-%d")
ax1.set_xlim([start_date, end_date])

ax1.legend(loc="upper left")
ax2.legend(loc="upper left", bbox_to_anchor=(0, 0.95))

plt.xticks()

plt.show()
```
